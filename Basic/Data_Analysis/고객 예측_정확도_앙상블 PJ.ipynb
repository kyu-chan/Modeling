```python
##범주형 데이터가 많은데 ONE HOT 인코딩으로 일단 처리한다
data.dtypes
```

```python
## y는 종속변수이니 건드리지말고 아래 범주형변수를 원핫인코딩합시다
## get_dummies사용
data = pd.get_dummies(data, columns=['job','marital','education','default','housing','loan','contact','month','day_of_week','poutcome'])
```

```python
data.head()
```

트레이닝/테스트 셋 구분

```python
data['id']= range(len(data))
```

```python
len(data)
```

```python
## 30000개  training 남는거 test 하자
## 나중에 비복원으로도 해봐
train = data.sample(30000, replace=False, random_state=2021).reset_index().drop(['index'],axis=1)
test = data.loc[ ~data['id'].isin(train['id'])].reset_index().drop(['index'],axis=1)
## 이 테스트, 트레이닝 기준을 나눌때 주의해야할 것이..
```

<strong> 해석이 어렵고 학습이 느리다는 문제 <br>
<strong> n_estimators = 몇개의 의사결정나무를 사용할 것인지 (보통 100~500) <br>
<strong> min_samples_split : 각 의사결정 나무에서 각 구간의 최소 샘플 수 , 이보다 작으면 더 이상 분리되지 않음

```python
from sklearn.ensemble import RandomForestClassifier

rf = RandomForestClassifier(n_estimators= 500, min_samples_split= 10)
```
